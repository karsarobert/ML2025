#Gridsearch

![](https://i0.wp.com/sqlrelease.com/wp-content/uploads/2021/08/An-introduction-to-GridSearchCV-and-RandomizedSearchCV-Image.jpg?w=954&ssl=1)

```python
from sklearn.model_selection import GridSearchCV # paraméterek hangolása
rf_clf = DecisionTreeClassifier() # a döntési fa osztályozó
params_grid = {"max_depth" : [3,4,5,6,7,8,9,10, None],
              "min_samples_split": [2, 3, 10, 20], # A min_samples_split paraméter azt határozza meg, hogy egy belső csomópontot (node) legalább hány minta esetén oszt fel a döntési fa.
              } # döntési fához tartozó hiperparaméterek felsorolása egy dictionary be
grid_search = GridSearchCV(rf_clf, params_grid,
                           n_jobs=-1, cv=5, scoring='accuracy') #grid search beállítása n_jobs paraméter azt határozza meg, hogy a grid search futtatása során hány párhuzamos folyamat (CPU mag) legyen használva. Ha -1-et adunk meg, akkor a rendszer az összes elérhető processzormagot kihasználja
grid_search.fit(X, y) #illesztés
```

```python
grid_search.best_params_ # a megadott paraméterekből a legjobb találat
```

```python
from sklearn.metrics import accuracy_score
cvrf_clf = grid_search.best_estimator_ # a legjobb modell kiválasztása
grid_search.best_score_ # a legjobb modell pontossága
```

#Ensemble (együttes) módszerek
##Majority voting, többségi szavazás (párhuzamosítható feladat)

![](https://www.researchgate.net/publication/324014302/figure/fig2/AS:644424015040514@1530654066950/Majority-voting-algorithm.png)

```python
from sklearn import model_selection
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn import datasets
from sklearn.ensemble import VotingClassifier # többségi szavazás
from sklearn.linear_model import LogisticRegression

import pandas as pd

df = pd.read_csv("https://raw.githubusercontent.com/karsarobert/ML2025/main/diabetes.csv") # adatok betöltése

y = df.label #X,y beállítása
X = df.drop(['label'], axis = 1)

X_train, X_test, y_train, y_test = \
    train_test_split(X, y, test_size=0.2, random_state=1) # első felosztás train, teszt halmazokra

X_train, X_val, y_train, y_val = \
    train_test_split(X_train, y_train, test_size=0.2, random_state=1) # második felosztás a train halmazt tovább osztjuk train és valid halmazokra

print('Train/Valid/Test sizes:', y_train.shape[0], y_val.shape[0], y_test.shape[0]) # az egyes halmazok elemszámainak kiiratása

# definiálunk 3 osztályozót
clf1 = DecisionTreeClassifier(random_state=1, max_depth=None)
# max_depth=None a csomópontok addig bővülnek, amíg az összes levél tiszta nem lesz, vagy amíg az összes levél kevesebb mint min_samples_split mintát tartalmaz.
clf2 = DecisionTreeClassifier(random_state=1, max_depth=5)
clf3 = LogisticRegression(solver='lbfgs', max_iter=1000)

clf1.fit(X_train, y_train) # az egyes modellek (clf1-clf3) illesztését csak azért végezzük el, hogy lássuk az eredményeket!
clf2.fit(X_train, y_train)
clf3.fit(X_train, y_train)
eclf = VotingClassifier(estimators=[   # együttes osztályozó létrehozása
        ('DTC1', clf1), ('DTC2', clf2), ('LR3', clf3)],
        voting='hard', weights=[1,1,1]) # voting paraméter lehet hard és soft
        # a soft voting nem a döntés szerinti osztályokkal dolgozik haenm a valószínűségi kimenetekkel!


eclf = eclf.fit(X_train, y_train) # együttes osztályozó illesztése
print("clf1 Accuracy: %0.4f" % clf1.score(X_val, y_val))
print("clf2 Accuracy: %0.4f" % clf2.score(X_val, y_val))
print("clf3 Accuracy: %0.4f" % clf3.score(X_val, y_val))
print("Ensemble Accuracy: %0.4f" % eclf.score(X_val, y_val))
```

#Bagging (bootstrap aggregating)

##Bootstrap sampling

A Bootstrap sampling egy statisztikai technika, amely során az eredeti adathalmazból visszatevéssel választunk véletlenszerűen mintákat. Ez azt jelenti, hogy egy adott minta akár többször is kiválasztható, míg mások kimaradnak az adott kiválasztásból. Ezzel több, egymástól kissé eltérő adatmintát kapunk, amelyeket például ensemble módszerek, mint a bagging esetén használnak a modellek varianciájának csökkentésére és a teljesítmény javítására.

![](http://rasbt.github.io/mlxtend/user_guide/evaluate/BootstrapOutOfBag_files/bootstrap_concept.png)


A Bagging (bootstrap aggregating) egy ensemble módszer, amely több modellt tanít az eredeti adathalmazból véletlenszerűen, visszatevéssel kiválasztott (bootstrap) minták alapján. Az egyes modellek által adott eredményeket aggregálva (például többségi szavazással osztályozásnál) csökkenthető a modell varianciája, javítva ezzel a stabilitást és a generalizációt, illetve csökkenthető a túltanulás esélye. A scikit-learn-ben a BaggingClassifier pontosan ezt a módszert valósítja meg, lehetővé téve több alaptanuló modell (például döntési fa) együttes alkalmazását a jobb teljesítmény érdekében.

![](https://static.packt-cdn.com/products/9781787125933/graphics/B07030_07_06.jpg)

![](https://datavedas.com/wp-content/uploads/2018/04/image003.jpg)

```python
from sklearn import model_selection
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn import datasets
from sklearn.ensemble import BaggingClassifier

import pandas as pd

df = pd.read_csv("https://raw.githubusercontent.com/karsarobert/ML2025/main/diabetes.csv") # adatok betöltése

y = df.label #X,y beállítása
X = df.drop(['label'], axis = 1)


X_train, X_test, y_train, y_test = \
    train_test_split(X, y, test_size=0.2, random_state=1) # első felosztás train, teszt halmazokra

X_train, X_val, y_train, y_val = \
    train_test_split(X_train, y_train, test_size=0.2, random_state=1) # második felosztás a train halmazt tovább osztjuk train és valid halmazokra

print('Train/Valid/Test sizes:', y_train.shape[0], y_val.shape[0], y_test.shape[0])

tree = DecisionTreeClassifier(criterion='entropy', # döntési fa osztályozó létrehozása
                              random_state=1,
                              max_depth=None)


bag = BaggingClassifier(estimator=tree, # bagging osztályozó létrehozása
                        n_estimators=500, # becslők száma n_estimators=500: az ensemble 500 különálló döntési fát épít fel. Minden egyes fa különböző bootstrap mintán tanul.
                        oob_score=True, # mérjük-e a teszt halmaz eredményeit
                        bootstrap=True, # bootstrap alkalmazása, ha a bootstrap=False, akkor a mintavételezés nem történik visszatevéssel, azaz minden alaptanuló modellhez az adatokból véletlenszerűen kiválasztott részhalmazt kapjuk úgy, hogy egy adott adatpont egyszer sem szerepelhet többször ugyanabban a részhalmazban.
                        n_jobs=1, # hány szálon fusson
                        random_state=1)

bag.fit(X_train, y_train)

print("OOB Accuracy: %0.2f" % bag.oob_score_) #zsákon kívüli adatokra (test set) vonatkozó pontosság!
print("Test Accuracy: %0.2f" % bag.score(X_test, y_test))
```

```python
tree = DecisionTreeClassifier(criterion='entropy', # egy fa pontossága
                              random_state=1,
                              max_depth=None)

tree.fit(X_train, y_train)

print("Test Accuracy: %0.2f" % tree.score(X_test, y_test))
```

#AdaBoosting

Az adaptív boost ing (más néven AdaBoost) az egyik leggyakrabban használt boosting ensemble módszer. Az alkalmazkodó boosting egyesíti ( fellendíti ) a gyenge tanulókat, és erős tanulóvá alakítja.

![](https://www.researchgate.net/profile/Mustafa-Fayez/publication/353826737/figure/fig4/AS:1055764118515712@1628725189980/4-Ada-Boost-classifier-steps-49.jpg)

Az 1. osztályozó osztályozza a pontokat (1. kép). Vannak olyan pontok (kék pontok), amelyeket rosszul osztályoznak.

A rosszul besorolt ​​pontok nagyobb súlyt kapnak, és a helyesen besorolt ​​pontok súlya csökken. Egy új osztályozó olyan új képzési adatkészlettel van betanítva, amely több súlyt rendel a rosszul osztályozott pontokhoz és kisebb súlyokat a helyesen osztályozott pontokhoz (így adaptív újramintavételezés ).

Nézze meg a 2. képet, ahol a rosszul besorolt ​​pontok nagyobb súlyt kaptak, ami nagyobb +-al vagy -al van ábrázolva, és a helyesen osztályozott pontok mérete csökkent. Az új osztályozó ismét néhány pont (3 kék pont) hibás besorolását eredményezi.

A 2. osztályozó által besorolt ​​rosszul besorolt ​​pontok nagyobb súlyt kapnak, a helyesen besorolt ​​pontok pedig kisebb súlyt kapnak (csökkentett méret).

A 3. osztályozó új képzési adatkészletekkel újraképződik, minden egyes adatpont súlyozásával. Vessen egy pillantást a 3. képre.

Végül az ensemble adaptív boosting osztályozó három osztályozó összeállításával készül, amelyeket az adaptív újramintavételezés eredményeként létrehozott különböző képzési adatkészletek felhasználásával állítottak össze / illesztettek / betanítottak. A 4. kép, amely az adaptív boost (AdaBoost) osztályozót ábrázolja, amelyet három osztályozó összeállításával jött létre.

```python
from sklearn import model_selection
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn import datasets
from sklearn.ensemble import AdaBoostClassifier

import pandas as pd

df = pd.read_csv("https://raw.githubusercontent.com/karsarobert/ML2025/main/diabetes.csv") # adatok betöltése

y = df.label #X,y beállítása
X = df.drop(['label'], axis = 1)

X_train, X_test, y_train, y_test = \
    train_test_split(X, y, test_size=0.25, random_state=1)

X_train, X_val, y_train, y_val = \
    train_test_split(X_train, y_train, test_size=0.25, random_state=1)

print('Train/Valid/Test sizes:', y_train.shape[0], y_val.shape[0], y_test.shape[0])

tree = DecisionTreeClassifier(criterion='entropy',
                              random_state=1,
                              max_depth=1)


boost = AdaBoostClassifier(estimator=tree,
                           n_estimators=500,
                           algorithm='SAMME',
                           random_state=1)

boost.fit(X_train, y_train)

print("Test Accuracy: %0.2f" % boost.score(X_test, y_test))
```

#Random forest classifier

A **Random Forest** (vagy véletlen erdő) egy úgynevezett „ensemble” tanulási módszer, amely sok döntési fát (osztályozókat) hoz létre, majd ezek eredményét kombinálva (például többségi szavazással) adja meg a végső kimenetet. Az eljárás célja a döntési fák esetleges túltanulását és magas varianciáját csökkenteni, így stabilabb, robusztusabb modellt kapunk.  

A **Random Forest** működésének főbb lépései:

1. **Bootstrap mintavételezés (Bagging) alkalmazása**  
   - Minden döntési fát az eredeti tanító adathalmazból bootstrap technikával (visszatevéses mintavétellel) vesznek fel.  
   - Így létrejön több, egymástól kismértékben eltérő tanító halmaz.  
   - Mivel a mintavételezés visszatevéses, ugyanaz az adatpont többször is bekerülhet egy-egy fa tanítóhalmazába, míg más pontok esetleg egyszer sem.

2. **Véletlen jellemző (feature) kiválasztás**  
   - A Random Forest nem csak az adatpontok kiválasztását randomizálja, hanem a jellemzők (változók) megválasztását is.  
   - Minden egyes csomópontban (split) nem az összes rendelkezésre álló jellemző közül választjuk ki a legjobbat, hanem egy véletlenszerűen kiválasztott, kisebb jellemzőhalmazból (például \(\sqrt{d}\) darab a \(d\) összes jellemző közül).  
   - Ezzel biztosítjuk, hogy a fák ne legyenek túlságosan hasonlóak egymáshoz (csökken a korrelációjuk).

3. **Döntési fák felépítése**  
   - Minden bootstrap mintán, az adott véletlenszerű jellemzőválasztást használva felépül egy-egy döntési fa.  
   - A fák növesztése addig tart, amíg valamilyen megadott feltételt el nem érnek (például minimális mintaszám a csomópontban, maximális fa-mélység, stb.).

4. **Ensemble (többségi szavazás) vagy átlagolás**  
   - A létrejött fák mindegyike önálló osztályozóként működik, de a végeredményt az összes fa predikciójának kombinálásával kapjuk meg.  
   - Osztályozási feladatnál jellemzően többségi szavazással dől el a végső címke, regressziós feladatnál pedig átlagoljuk a fák becsléseit.

5. **OOB (Out-of-Bag) hiba és egyéb jellemzők**  
   - A bootstrap mintákból mindig kimarad néhány adatpont („zsákon kívüli” adatok). Ezekkel a kimaradt pontokkal minden egyes fa teljesítménye tesztelhető anélkül, hogy külön validációs halmazt tartanánk fenn.  
   - Ezáltal kapunk egy OOB hibabecslést, ami általában jó közelítése a valódi generalizációs hibának.

**Miért hatékony?**  
- Azáltal, hogy a modellek (fák) különböző részhalmazokon és véletlenszerű jellemzőválasztással tanulnak, a kapott fák eltérő hibákat vétenek.  
- A hibák átlagolása (vagy többségi szavazása) csökkenti a varianciát és általában jobb általánosító képességet eredményez, mintha csak egyetlen, akár nagyon mély döntési fát használunk.  
- Ellenállóbb a zajjal szemben és kisebb a túltanulás esélye.

Összefoglalva, a **Random Forest** a bagging és a véletlenszerű jellemző-kiválasztás együttes erejét használja ki: rengeteg „kicsit” különböző döntési fát hoz létre, majd ezek döntéseit egyesítve ad egy stabil, jól általánosító modellt.
